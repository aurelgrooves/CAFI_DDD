#### merge multiple CEO files and combine columns

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
# Set the display option to a large value to prevent text wrapping
pd.set_option('display.max_colwidth', None)
```

```python
# read clean ceo data from script 05d
ceo_gab = pd.read_csv('/home/sepal-user/module_results/esbae/GAB/GAB_all_ceo_1522_clean.csv')
ceo_drc = pd.read_csv('/home/sepal-user/module_results/esbae/DRC/DRC_all_ceo_1522_clean.csv')
ceo_eqg = pd.read_csv('/home/sepal-user/module_results/esbae/EQG/EQG_all_ceo_1522_clean.csv')
ceo_cog = pd.read_csv('/home/sepal-user/module_results/esbae/COG/COG_all_ceo_1522_clean.csv')
ceo_cog2 = pd.read_csv('/home/sepal-user/module_results/esbae/COG/COG_all_ceo_1622_clean.csv')
```

```python
data_frames = [ceo_gab, ceo_drc, ceo_eqg, ceo_cog, ceo_cog2] 
```

```python
# Source file names corresponding to each data frame
source_files = ['GAB', 'DRC', 'EQG','COG_1522', 'COG_16_22']  # Replace with your file names

# Merge data frames and add a new column for source file name
merged_ceo = pd.concat([df.assign(Source=source_file) for df, source_file in zip(data_frames, source_files)], ignore_index=True)
```

```python
# Print or use the merged data frame
merged_ceo.head()
```

```python
column_names = merged_ceo.columns.tolist()
# Print the list of column names
print(column_names)
```

```python
# do some clean up
merged_ceo = merged_ceo.drop(['kmeans', 'cnc_1520'], axis=1)
merged_ceo.rename(columns={'Unnamed: 0': 'index'}, inplace=True)
```

```python
# create new ref_change and ref years
merged_ceo['Ref_Change_Type'] = merged_ceo['Ref_Change_Type_1522'].fillna(merged_ceo['Ref_Change_Type_1622'])
merged_ceo['Ref_Change_Year'] = merged_ceo['Ref_Change_Year_1522'].fillna(merged_ceo['Ref_Change_Year_1622'])
```

```python
points_strata = pd.pivot_table(merged_ceo, values = 'point_id', index = 'Ref_Change_Type',columns = 'Source', aggfunc =  "count")
points_strata
```
